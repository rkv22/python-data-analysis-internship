```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Set visualization style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

print("All libraries imported successfully!")

# Load the dataset
df_math = pd.read_csv('student-mat.csv', sep=';')

print("Dataset loaded successfully!")
print(f"Shape: {df_math.shape}")
print(f"\nFirst 5 rows:")
df_math.head()

# Check dataset info
print("Dataset Information:")
print("="*50)
df_math.info()


# Check for missing values
print("\nMissing Values:")
print("="*50)
missing = df_math.isnull().sum()
print(missing[missing > 0] if missing.sum() > 0 else "No missing values found!")
print(f"\nTotal missing values: {df_math.isnull().sum().sum()}")

# Checking for duplicates
print("\nDuplicate Rows:")
print("="*50)
duplicates = df_math.duplicated().sum()
print(f"Total duplicate rows: {duplicates}")

if duplicates > 0:
    print(f"\nRemoving {duplicates} duplicate rows...")
    df_math = df_math.drop_duplicates()
    print(f"Dataset shape after removing duplicates: {df_math.shape}")


# Display basic statistics
print("\nBasic Statistics:")
print("="*50)
df_math.describe()

# Calculating average final grade
avg_grade = df_math['G3'].mean()
median_grade = df_math['G3'].median()
std_grade = df_math['G3'].std()

print("Final Grade Statistics (G3):")
print("="*50)
print(f"Average Final Grade: {avg_grade:.2f}")
print(f"Median Final Grade: {median_grade:.2f}")
print(f"Standard Deviation: {std_grade:.2f}")
print(f"Min Grade: {df_math['G3'].min()}")
print(f"Max Grade: {df_math['G3'].max()}")


# Counting students with grades above 15
above_15 = (df_math['G3'] > 15).sum()
percentage_above_15 = (above_15 / len(df_math)) * 100

print("Students Scoring Above 15:")
print("="*50)
print(f"Number of students with G3 > 15: {above_15}")
print(f"Percentage: {percentage_above_15:.2f}%")
print(f"Total students: {len(df_math)}")


# Calculate correlation between study time and grades
correlation = df_math['studytime'].corr(df_math['G3'])

print("Study Time vs Final Grade Correlation:")
print("="*50)
print(f"Pearson Correlation Coefficient: {correlation:.4f}")

# Perform statistical test
r, p_value = stats.pearsonr(df_math['studytime'], df_math['G3'])
print(f"P-value: {p_value:.6f}")

if p_value < 0.05:
    print("\n✓ The correlation is statistically significant (p < 0.05)")
else:
    print("\n✗ The correlation is NOT statistically significant (p ≥ 0.05)")

# Interpretation
if abs(correlation) < 0.3:
    strength = "weak"
elif abs(correlation) < 0.7:
    strength = "moderate"
else:
    strength = "strong"

direction = "positive" if correlation > 0 else "negative"
print(f"\nInterpretation: There is a {strength} {direction} relationship between study time and grades.")

# Average grade by study time
print("\nAverage Grade by Study Time Level:")
print("="*50)
studytime_labels = {1: '<2 hours', 2: '2-5 hours', 3: '5-10 hours', 4: '>10 hours'}

avg_by_studytime = df_math.groupby('studytime')['G3'].agg(['mean', 'count', 'std']).round(2)
avg_by_studytime.index = [studytime_labels.get(i, i) for i in avg_by_studytime.index]
print(avg_by_studytime)


# Compare performance by gender
print("Performance by Gender:")
print("="*50)

gender_stats = df_math.groupby('sex')['G3'].agg(['count', 'mean', 'median', 'std']).round(2)
gender_stats.index = ['Female' if i == 'F' else 'Male' for i in gender_stats.index]
print(gender_stats)

# Calculate the difference
female_avg = df_math[df_math['sex'] == 'F']['G3'].mean()
male_avg = df_math[df_math['sex'] == 'M']['G3'].mean()
difference = female_avg - male_avg

print(f"\nAverage Grade Difference (Female - Male): {difference:.2f}")

if difference > 0:
    print(f"Females score {abs(difference):.2f} points higher on average.")
else:
    print(f"Males score {abs(difference):.2f} points higher on average.")

# Statistical test for gender differences
female_grades = df_math[df_math['sex'] == 'F']['G3']
male_grades = df_math[df_math['sex'] == 'M']['G3']

t_stat, p_value = stats.ttest_ind(female_grades, male_grades)

print("\nT-Test Results (Female vs Male):")
print("="*50)
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.6f}")

if p_value < 0.05:
    print("\n✓ The difference is statistically significant (p < 0.05)")
else:
    print("\n✗ The difference is NOT statistically significant (p ≥ 0.05)")


#Data Visualizations

# Histogram Of Final Grades
plt.figure(figsize=(12, 6))
plt.hist(df_math['G3'], bins=20, color='steelblue', edgecolor='black', alpha=0.7)
plt.axvline(avg_grade, color='red', linestyle='--', linewidth=2, label=f'Mean: {avg_grade:.2f}')
plt.axvline(median_grade, color='green', linestyle='--', linewidth=2, label=f'Median: {median_grade:.2f}')

plt.xlabel('Final Grade (G3)', fontsize=12, fontweight='bold')
plt.ylabel('Frequency', fontsize=12, fontweight='bold')
plt.title('Distribution of Final Grades in Mathematics', fontsize=14, fontweight='bold')
plt.legend(fontsize=10)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print("Histogram of Final Grades displayed.")

plt.figure(figsize=(12, 6))

# Create scatter plot with jitter for better visibility
jitter_x = df_math['studytime'] + np.random.normal(0, 0.04, size=len(df_math))
plt.scatter(jitter_x, df_math['G3'], alpha=0.5, s=50, color='steelblue')

# Add trend line
z = np.polyfit(df_math['studytime'], df_math['G3'], 1)
p = np.poly1d(z)
plt.plot(df_math['studytime'].unique(), p(df_math['studytime'].unique()),
         "r--", linewidth=2, label=f'Trend Line (r={correlation:.3f})')

plt.xlabel('Study Time Level', fontsize=12, fontweight='bold')
plt.ylabel('Final Grade (G3)', fontsize=12, fontweight='bold')
plt.title('Relationship between Study Time and Final Grade', fontsize=14, fontweight='bold')
plt.xticks([1, 2, 3, 4], ['<2 hrs', '2-5 hrs', '5-10 hrs', '>10 hrs'])
plt.legend(fontsize=10)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

print("Scatter plot displayed.")

ig, axes = plt.subplots(1, 2, figsize=(14, 5))

#Bar Graph

# Average score by gender
gender_avg = df_math.groupby('sex')['G3'].mean()
gender_labels = ['Female', 'Male']
colors = ['#FF6B9D', '#4A90E2']

axes[0].bar(gender_labels, gender_avg.values, color=colors, alpha=0.7, edgecolor='black', linewidth=1.5)
axes[0].set_ylabel('Average Grade', fontsize=11, fontweight='bold')
axes[0].set_title('Average Final Grade by Gender', fontsize=12, fontweight='bold')
axes[0].set_ylim([0, 20])
axes[0].grid(axis='y', alpha=0.3)

# Add value labels on bars
for i, v in enumerate(gender_avg.values):
    axes[0].text(i, v + 0.3, f'{v:.2f}', ha='center', fontweight='bold')

# Student count by gender
gender_count = df_math['sex'].value_counts()
axes[1].bar(gender_labels, gender_count.values, color=colors, alpha=0.7, edgecolor='black', linewidth=1.5)
axes[1].set_ylabel('Number of Students', fontsize=11, fontweight='bold')
axes[1].set_title('Student Distribution by Gender', fontsize=12, fontweight='bold')
axes[1].grid(axis='y', alpha=0.3)

# Add value labels on bars
for i, v in enumerate(gender_count.values):
    axes[1].text(i, v + 5, str(v), ha='center', fontweight='bold')

plt.tight_layout()
plt.show()

print("Gender comparison charts displayed.")

```
